```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.pandas.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
# Xác định biến phân loại có giá trị thiếu
features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']
for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
#thay thế giá trị bị thiếu
def replace(df,features_nan):
    data=df.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

df=replace(df,features_nan)
df[features_nan].isnull().sum()
```

```python
df.head()
```

```python
#biến số bị thiếu
numerical_with_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
#thay thế giá trị bị thiếu bằng mean 
for feature in numerical_with_nan:
    median_value=df[feature].median()

    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
    
df[numerical_with_nan].isnull().sum()
```

```python
df.head()
```

```python
#đưa các biến thành tuổi
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    df[feature]=df['YrSold']-df[feature]
```

```python
df.head()
```

```python
df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
#logarit hoá cho các biến số
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

```python
#tìm biến phân loại
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
categorical_features
```

```python
#tính tuần suất 
for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
    
```

```python
df.head()
```

```python
#sắp xếp lại dựa trên mean của 'SalePrice'
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
df.head(10)
```

```python
scaling_feature=[feature for feature in df.columns if feature not in ['Id','SalePerice'] ]
len(scaling_feature)
feature_scale=[feature for feature in df.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[feature_scale])
```

```python
scaler.transform(df[feature_scale])
```

```python
data = pd.concat([df[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(df[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data.head()
```

```python
features = data[['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']]
target = data['SalePrice']
```

```python
scaled_features = StandardScaler().fit_transform(features)
X_train, X_test, y_train, y_test = train_test_split(scaled_features, target, test_size=0.2, random_state=42)
```

```python
max_k = 20
mse_scores = [mean_squared_error(y_test, KNeighborsRegressor(n_neighbors=k).fit(X_train, y_train).predict(X_test)) for k in range(1, max_k+1)]
optimal_k = mse_scores.index(min(mse_scores)) + 1
knn = KNeighborsRegressor(n_neighbors=optimal_k).fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
result_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(result_df)
```

```python
mse_test = mean_squared_error(y_test, y_pred)
r2_test = r2_score(y_test, y_pred)
print(f'Mean Squared Error on Test Set: {mse_test}')
print(f'R-squared on Test Set: {r2_test}')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_k+1), mse_scores, marker='o')
plt.title('Mean Squared Error for Different Values of K')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Mean Squared Error')
plt.show()
```
